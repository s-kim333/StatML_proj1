```python
df_shuffled=df_shuffled.iloc[np.random.permutation(df_shuffled.index)].reset_index(drop=True)
```

```python
train_df, valid_df = train_test_split(df_shuffled, test_size=0.3, random_state=7)
```

```python
X_train = train_df.drop("class", axis=1)
Y_train = train_df["class"]
X_valid = valid_df.drop("class", axis=1)
Y_valid = valid_df["class"]
X_valid
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
sc = StandardScaler()

features = ["Node_1_friends_number","Node_2_friends_number","number_of_common_friends","sum_of_friends"]

X_train = sc.fit_transform(train_df[features]) 

X_valid = sc.fit_transform(valid_df[features])
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0,probability = True)
classifier.fit(X_train, Y_train)
Y_pred = classifier.predict_proba(X_valid)

from sklearn.metrics import roc_auc_score as score
classifier.score(X_train, Y_train)
classifier = round(classifier.score(X_train, Y_train) * 100, 2)
classifier
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, Y_train)
Y_pred = classifier.predict_proba(X_valid)

from sklearn.metrics import roc_auc_score as score
classifier.score(X_train, Y_train)
classifier = round(classifier.score(X_train, Y_train) * 100, 2)
classifier
```
